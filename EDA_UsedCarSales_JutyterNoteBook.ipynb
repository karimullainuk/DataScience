### **Sales and Revenues of each segment of class** 

````python
'''
Fnction that computes the class of the car based on price
Logic:
       Price Range               Class
       -------------             -------------
       <= 10000                  Economy
       > 10000 and <= 30000      Luxury
       > 30000                   Premium

'''
def car_class(pricedf):
    if pricedf.price <= 10000:
        return "Economy"
    elif (pricedf.price > 10000) & (pricedf.price <=30000):
        return "Luxury"
    else:
        return "Premium"
````

````python
# Add new column 'class' for defining the segment
data_new['class'] = data_new.apply(car_class, axis=1)
````
